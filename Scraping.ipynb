```python
import requests
import pandas
from bs4 import BeautifulSoup

#Make a dictionary with classes for each attribute
property_dict = { 
    "dict_prices" : [],
    "dict_address" : [],
    "dict_sqft" :[],
    "dict_acres" : [],
    "dict_url" : [],
    "dict_broker" : [],
}

pgnumber=1

while(True):

    # URL of the page we want to scrape.
    url = f"https://www.realtor.com/realestateandhomes-search/Brownwood_TX/pg-{pgnumber}"

    # Headers are used to make the request look like it is coming from a browser.
    HEADERS = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36",
        "accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
        "accept-language": "en-US;en;q=0.9",
        "accept-encoding": "gzip, deflate, br",
    }
    # Send an HTTP request to the URL.
    response = requests.get(url, headers = HEADERS)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.text, 'html.parser')
        if soup.find('div', class_='BasePropertyCard_propertyCardWrap__Z5y4p') == None:
            break
        
        print("Check Point")
        print(pgnumber)

        for price_divs in soup.find_all('div', class_='Pricestyles__StyledPrice-rui__btk3ge-0 bvgLFe card-price'):
            property_dict["dict_prices"].append(price_divs.get_text(strip=True))
        
        for address_div in soup.find_all('div', class_='card-address truncate-line'):
            # Extract the text content of both address lines
            if address_div.find('div', {'data-testid': 'card-address-1'}) is None:
                address_line_1 = 'N/A'
            else:
                address_line_1 = address_div.find('div', {'data-testid': 'card-address-1'}).get_text(strip=True)
            if address_div.find('div', {'data-testid': 'card-address-2'}) is None:
                address_line_2 = 'N/A'
            else:
                address_line_2 = address_div.find('div', {'data-testid': 'card-address-2'}).get_text(strip=True)

            # Combine the address lines to get the complete address
            complete_address = f"{address_line_1}, {address_line_2}"

            # Append the complete address to the list
            property_dict["dict_address"].append(complete_address)

        for url_div in soup.find_all('div', class_='CardContent__StyledCardContent-rui__sc-7ptz1z-0 dsJFdl card-content card-content'):
            a_tag = url_div.find('a', class_='LinkComponent_anchor__0C2xC')
            relative_url = a_tag.get('href')
            complete_url = f"https://www.realtor.com{relative_url}"
            property_dict["dict_url"].append(complete_url)

        for span_tag in soup.find_all('span', class_='BrokerTitle_titleText__Y8pb0'):
            property_dict["dict_broker"].append(span_tag.get_text(strip=True))

        for property_div in soup.find_all('ul', class_='PropertyMetastyles__StyledPropertyMeta-rui__sc-1g5rdjn-0 KKDDp card-meta'):
            # Find the <li> tag with the specific data-testid attribute for sqft size
            sqft_li = property_div.find('li', class_='PropertySqftMetastyles__StyledPropertySqftMeta-rui__sc-1gdau7i-0 fnhaOV')
            # Extract the sqft size value
            if sqft_li is None:
                sqft_size = 'N/A'
            else:
                sqft_size = sqft_li.find('span', class_='meta-value').get_text(strip=True)
            
            # Append the sqft size to the list
            property_dict["dict_sqft"].append(sqft_size)    
        
        for property_div in soup.find_all('ul', class_='PropertyMetastyles__StyledPropertyMeta-rui__sc-1g5rdjn-0 KKDDp card-meta'):
            # Find the <li> tag with the specific data-testid attribute for acre lot size
            acre_lot_li = property_div.find('li', class_='PropertyLotSizeMetastyles__StyledPropertyLotSizeMeta-rui__sc-1cz4zco-0 cNMyen')
            # Extract the acre lot size value
            if acre_lot_li is None:
                acre_lot_size = 'N/A'
            else:
                acre_lot_size = acre_lot_li.find('span', class_='meta-value').get_text(strip=True)
            
            # Append the acre lot size to the list
            property_dict["dict_acres"].append(acre_lot_size)

        pgnumber += 1
    elif(response.status_code == 404):
        print(f"Failed to retrieve the page. Status code: {response.status_code}")
        break
```
